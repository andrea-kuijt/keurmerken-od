```
import os
from google.colab import drive, files
import imageai
from imageai.Detection.Custom import CustomObjectDetection
```

```
drive.mount('/content/drive/')
```

```
detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("/content/drive/Shared drives/Class 15/keurmerken-od-training/planetproof/models/detection_model-ex-027--loss-0003.288.h5")
detector.setJsonPath("/content/drive/Shared drives/Class 15/keurmerken-od-predicting/planetproof/detection_config.json")
detector.loadModel()
```

```
import os

image_dir = '/content/drive/Shared drives/Class 15/keurmerken-od-predicting/planetproof/test/images/'
results_dir = '/content/drive/Shared drives/Class 15/keurmerken-od-predicting/planetproof/test/results/'

for image_file_name in os.listdir('/content/drive/Shared drives/Class 15/keurmerken-od-predicting/planetproof/test/images'):
    print('\n'+image_file_name)
    detections = detector.detectObjectsFromImage(
        input_image = image_dir + image_file_name,
        output_image_path = results_dir + image_file_name,
        minimum_percentage_probability=30)
    
    for detection in detections:
        print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
```
